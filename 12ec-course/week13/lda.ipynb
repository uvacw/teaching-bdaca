```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import gensim
from gensim import matutils, corpora
from gensim.models.ldamodel import LdaModel
from gensim.models.coherencemodel import CoherenceModel
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis
```

```python
url = "https://cssbook.net/d/sotu.csv"
sotu = pd.read_csv(url)
p_obama = (sotu[sotu.President == "Obama"].text.str.split("\n\n").explode())
```

```python
cv = TfidfVectorizer(min_df=.01, stop_words="english", ngram_range=(1,2))
dtm = cv.fit_transform(p_obama)
dtm
```

```python
corpus = matutils.Sparse2Corpus(dtm, documents_columns=False)
vocab = corpora.Dictionary()
vocab.id2token = dict(enumerate(cv.get_feature_names()))
vocab.token2id = dict((k, v) for k, v in cv.vocabulary_.items())
```

```python
lda = LdaModel(corpus,id2word=vocab,num_topics=10, random_state=123, alpha="asymmetric")

# same thing as looping over lda.print_topics but with a data frame
pd.DataFrame({f"Topic {n}":[w for (w,tw) in words]
    for (n, words) in 
    lda.show_topics(formatted=False)})
```

```python
topics = pd.DataFrame([dict(lda.get_document_topics(doc, 
          minimum_probability=0.0))
     for doc in corpus])
meta = (sotu.iloc[p_obama.index]
        .drop(columns=["text"])
        .reset_index(drop=True))
tpd = pd.concat([meta, topics], axis=1)
tpd.head()
```

```python
vis_data = gensimvis.prepare(lda, corpus, vocab)
pyLDAvis.display(vis_data)
```
